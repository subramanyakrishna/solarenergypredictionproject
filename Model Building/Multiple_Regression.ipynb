```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data_daily = pd.read_csv("hourly-without-night-hrs.csv")
data_daily.head()
```

```python
# create a scaler object
scaler = MinMaxScaler()
#scalar = StandardScaler()

df_norm = pd.DataFrame(scaler.fit_transform(data_daily.iloc[:,1:]), columns=data_daily.columns[1:])

#df_norm["Solar Energy"] = list(data_daily.iloc[:,-1])
```

```python
train, test = train_test_split(data_daily,  test_size=0.3)
train
```

```python
test
```

```python
reg = LinearRegression()
reg.fit(train.iloc[:,1:8],train.iloc[:,-1])
print("Multiple Regression Coefficients are: " + str(reg.coef_), end = "\n\n")
print("Intercept: ", str(reg.intercept_), end = "\n\n")
```

```python
pred = reg.predict(test.iloc[:,1:8])

# The mean squared error
print('Mean squared error: %.2f' % mean_squared_error(test.iloc[:,-1], pred))

# The coefficient of determination
print('Coefficient of determination: %.2f' % r2_score(test.iloc[:,-1], pred))
```
